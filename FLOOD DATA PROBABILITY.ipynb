## Modelling

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score,confusion_matrix,roc_curve
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import VotingClassifier
```

```python
enc = LabelEncoder()
```

```python
flood['FLOODS'] = enc.fit_transform(flood['FLOODS'])
```

```python
flood.columns
```

```python
x = flood.drop(columns=['SUBDIVISION','FLOODS'])
Y = flood['FLOODS']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,Y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
models = [LogisticRegression(max_iter=1000),SVC(kernel='linear'),KNeighborsClassifier(),RandomForestClassifier(),DecisionTreeClassifier()]
```

```python
def compare_models():
    for model in models:
        
        #train
        model.fit(x_train,y_train)
        
        #prediction
        y_pred = model.predict(x_test)
        
        accuracy = accuracy_score(y_test,y_pred)
        f1 = f1_score(y_test,y_pred)
        pre = precision_score(y_test,y_pred)
        recall = recall_score(y_test,y_pred)
        
        
        print("Accuracy Score of the ",model, ' = ',accuracy)
        print("f1 Score of the ",model, ' = ',f1)
        print("Precision Score of the ",model, ' = ',pre)
        print("Recall Score of the ",model, ' = ',recall)
```

```python
compare_models()
```

```python
cv_score_rand = cross_val_score(RandomForestClassifier(),x,Y,cv=5)
```

```python
cv_score_rand
```

```python
mean_acc_rand = sum(cv_score_rand)/5
per = mean_acc_rand * 100
round(per,2)
```

```python
def model_cross_validation():
    for model in models:
        cv_score = cross_val_score(model,x,Y,cv=5)
        mean_acc = (sum(cv_score)/5)*100
        mean_acc = round(mean_acc,2)

        print("Cross validation for ",model , "= ", cv_score)
        print("mean accureacy ",model, " = ", mean_acc)
```

```python
model_cross_validation()
```

RandomForest and Decision Tree has the most mean accuracy 

##### RandomForest

```python
rmf = RandomForestClassifier(max_depth=3,random_state=0)
rmf_clf = rmf.fit(x_train,y_train)
rmf_clf
```

```python
y_pred = rmf_clf.predict(x_test)
y_pred
```

```python
# Calculate recall score
recall = recall_score(y_test, y_pred) * 100
print("Recall score: {:.2f}%".format(recall))

# Calculate ROC AUC score
roc_auc = roc_auc_score(y_test, y_pred) * 100
print("ROC AUC score: {:.2f}%".format(roc_auc))

# Calculate precision
precision = precision_score(y_test, y_pred)
print(f'Precision: {precision:.2f}')

# Create confusion matrix
con_mat = confusion_matrix(y_test, y_pred)

# Plot ROC AUC curve (assuming you have the predicted probabilities)
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
plt.figure(figsize=(5,4))
plt.plot(fpr, tpr, label='ROC Curve (AUC: {:.2f}%)'.format(roc_auc))
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(0, 1.05)
plt.show()

# Plot confusion matrix heatmap
plt.figure(figsize=(4,3))
print('confusion metrix is:\n',con_mat)
sns.heatmap(con_mat, annot=True, fmt='d', cmap='viridis')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

#### Decision Tree

```python
dt = DecisionTreeClassifier(max_depth=7,random_state=0)
dt_clf = dt.fit(x_train,y_train)
dt_clf
```

```python
predict_tree = dt_clf.predict(x_test)
predict_tree
```

```python
# Calculate recall score
recall = recall_score(y_test, predict_tree) * 100
print("Recall score: {:.2f}%".format(recall))

# Calculate ROC AUC score
roc_auc = roc_auc_score(y_test, predict_tree) * 100
print("ROC AUC score: {:.2f}%".format(roc_auc))

# Calculate precision
precision = precision_score(y_test, predict_tree)
print(f'Precision: {precision:.2f}')

# Create confusion matrix
con_mat = confusion_matrix(y_test, predict_tree)

# Plot confusion matrix heatmap
plt.figure(figsize=(4,3))
print('confusion metrix is:\n',con_mat)
sns.heatmap(con_mat, annot=True, fmt='d', cmap='viridis')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
